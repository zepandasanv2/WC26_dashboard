# Notebook d'exploration

## 1. Lecture brute de la donnée

```python
from pathlib import Path
import pandas as pd

# Chemin absolu vers les données
project_root = Path().resolve()
matchs_path = project_root / "data" / "matchs.csv"
joueurs_path = project_root / "data" / "joueurs.csv"

# Charger les données
matchs = pd.read_csv(matchs_path)
joueurs = pd.read_csv(joueurs_path)

# Afficher les premières lignes
display(matchs.head())
display(joueurs.head())
```

## 2. En graphique

```python
import matplotlib.pyplot as plt
import seaborn as sns

# --- Graphique 1 : Top buteurs ---
top_buteurs = joueurs.sort_values(by="buts", ascending=False).head(5)
plt.figure(figsize=(10, 5))
sns.barplot(x="nom", y="buts", data=top_buteurs, palette="viridis")
plt.title("Top 5 des buteurs - CDM 2026")
plt.xlabel("Joueurs")
plt.ylabel("Nombre de buts")
plt.show()

# --- Graphique 2 : Buts par équipe ---
# Calculer le nombre de buts par équipe (à partir des joueurs)
buts_par_equipe = joueurs.groupby("equipe")["buts"].sum().sort_values(ascending=False)
plt.figure(figsize=(10, 5))
sns.barplot(x=buts_par_equipe.index, y=buts_par_equipe.values, palette="rocket")
plt.title("Nombre de buts par équipe - CDM 2026")
plt.xlabel("Équipes")
plt.ylabel("Nombre de buts")
plt.xticks(rotation=45)
plt.show()

# --- Graphique 3 : Répartition des cartes jaunes ---
plt.figure(figsize=(8, 5))
sns.countplot(x="cartons_jaunes", data=joueurs, palette="mako")
plt.title("Répartition des cartons jaunes - CDM 2026")
plt.xlabel("Nombre de cartons jaunes")
plt.ylabel("Nombre de joueurs")
plt.show()
```
